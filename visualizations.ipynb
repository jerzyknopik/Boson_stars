```python
imeigenvalues2=np.imag(eigenvalues2)
reeigenvalues2=np.real(eigenvalues2)
imeigenvalues21=np.imag(eigenvalues21)
reeigenvalues21=np.real(eigenvalues21)
imeigenvalues2.sort(axis=1)
reeigenvalues2.sort(axis=1)
imeigenvalues21.sort(axis=1)
reeigenvalues21.sort(axis=1)
```

```python
a=plt.plot(omegagrid2,imeigenvalues2)
b=plt.plot(omegagrid21,imeigenvalues21)
plt.show()
```

```python
a=plt.plot(omegagrid2,reeigenvalues2)
b=plt.plot(omegagrid21,reeigenvalues21)
plt.show()
```

```python
eigenvalues3=np.append(eigenvalues2,eigenvalues21)
eigenvalues3=eigenvalues3.reshape(800,12)
omegagrid3=np.append(omegagrid2,omegagrid21)
eigenvalues3.shape
```

```python
imeigenvalues3=np.imag(eigenvalues3)
reeigenvalues3=np.real(eigenvalues3)
imeigenvalues3.sort(axis=1)
reeigenvalues3.sort(axis=1)
```

```python

a=plt.plot(omegagrid3,imeigenvalues3)
# plt.title(r'Imaginary part of first eigenvalues for $\alpha=0.15$')
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\mathfrak{I}(\lambda)$')
plt.savefig("imgroundstate1500.pdf")
plt.show()

```
